```
import pandas as pd

# For some reason this doesn't have header. 
# Column details given in separate pdf
epraccur_raw = pd.read_csv(
    'raw_data/epraccur.csv', 
    header=None,
    usecols=[0, 1, 4, 9, 10, 11]
    )
```

```
epraccur_raw.columns = ['gp_code', 'name', 'address_1', 'postcode', 'valid_from', 'valid_to']
epraccur_raw.head()
```

```
epraccur_raw_active_only = epraccur_raw[epraccur_raw['valid_to'].isna()]
```

```
epraccur_raw_active_only.head()
```

```
print(f"Before filtering: {len(epraccur_raw)}")
print(f"After filtering to active only: {len(epraccur_raw_active_only)}")
```

```
import requests
import pandas as pd

postcode_list = epraccur_raw_active_only['postcode'].unique()
print(len(postcode_list))
```

```
import numpy as np
```

```
i = 0
postcode_list[i:(i+1)*100]
```

```
i = 1
postcode_list[i:(i+1)*100].tolist()
```

```
range(int(np.ceil(len(postcode_list)/100)))
```

```
results = []

for i in range(int(np.ceil(len(postcode_list)/100))):

    postcode_list_filtered = postcode_list[i*100:(((i+1)*100)-1)]
    print(f"Collecting postcodes {i*100} to {(((i+1)*100)-1)}")

    # Send API request, passing in your postcodes as a list
    postcode_lookup = requests.post(
        "https://api.postcodes.io/postcodes", 
        json={"postcodes": postcode_list_filtered.tolist()}
    )

    results.append(postcode_lookup)


```

```
results_dfs = [pd.json_normalize(i.json()['result'], sep='_') 
for i in results]
```

```
results_dfs[0]
```

```
postcodes_df = pd.concat(results_dfs)
```

```
# join geographic data to surgeries

epraccur_final = epraccur_raw_active_only.merge(
    postcodes_df, how="left", 
    left_on="postcode", right_on="query"
    )
```

```
epraccur_final.to_csv('data/gp_surgery_locations.csv')
```
